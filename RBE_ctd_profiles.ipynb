```python
import pandas as pd
from matplotlib.ticker import MaxNLocator
%matplotlib qt5
from matplotlib import pyplot as plt

plt.rcParams.update({'font.size': 10,'font.family':'serif','font.serif':'Arial'})
```

```python
data7 = 'RBE_ctd_7_oct.xlsx'
df7 = pd.read_excel(data7)
```

```python
data11 = 'RBE_ctd_11_okt.xlsx'
df11 = pd.read_excel(data11)
```

```python
df11.describe().T
```

```python
df = pd.concat([df7, df11], axis=0)
df.describe().T
```

```python
df
```

```python
from matplotlib import pyplot as plt

plt.plot(df['Temperature'], df['Sea pressure'],'r')
plt.plot(df['Salinity'], df['Sea pressure'],'g')

#plt.plot(profile['TEMP2'], profile['PRES'],'g')
plt.gca().invert_yaxis()
plt.xlabel('Temperature [degC]')
plt.ylabel('Pressure [dbar]')
```

```python
from matplotlib import pyplot as plt

# Dados de exemplo
temperature = df['Temperature']
salinity = df['Salinity']
sea_pressure = df['Sea pressure']

# Crie a figura e os eixos
fig, ax1 = plt.subplots(figsize=(8, 6))

# Plote a temperatura no eixo Y invertido
ax1.plot(temperature, sea_pressure, 'r', label='Temperature [degC]')
ax1.set_xlabel('Temperature [degC]')
ax1.set_ylabel('Pressure [dbar]')
ax1.invert_yaxis()  # Inverta o eixo Y para que a pressão seja crescente para baixo

# Crie um segundo eixo compartilhando o mesmo eixo X
ax2 = ax1.twiny()

# Plote a salinidade no eixo X superior
ax2.plot(salinity, sea_pressure, 'g', label='Salinity')
ax2.set_xlabel('Salinity')

# Ajuste os limites dos eixos conforme necessário
ax1.set_xlim(13, 23)  # Limite do eixo X inferior
ax2.set_xlim(30, 33)  # Limite do eixo X superior


plt.title('Temperature and Salinity vs. Pressure')
plt.tight_layout()
plt.show()
```

```python
#each profile, just downcasting
# Defina os limites do intervalo de datas
S1_ti = pd.to_datetime('2023-10-07 09:03:20.000')
S1_tf = pd.to_datetime('2023-10-07 09:03:53.667')
```

```python
# Defina os limites do intervalo de datas
S2_ti = pd.to_datetime('2023-10-11  08:30:15.083')
S2_tf = pd.to_datetime('2023-10-11 08:31:33.333')
```

```python
# Filtrar os dados com base no intervalo de datas
S1 = df[(df['Time'] >= S1_ti) & (df['Time'] <= S1_tf)]
S2 = df[(df['Time'] >= S2_ti) & (df['Time'] <= S2_tf)]
# Agora, 'dados_no_intervalo' conterá apenas as linhas com datas dentro do intervalo especific
S2.describe()
```

```python
from matplotlib import pyplot as plt

plt.plot(S1['Temperature'], S1['Sea pressure'],'r', linestyle='--')
plt.plot(S1['Salinity'], S1['Sea pressure'],'g', linestyle='--')

plt.plot(S2['Temperature'], S2['Sea pressure'],'r')
plt.plot(S2['Salinity'], S2['Sea pressure'],'g')

#plt.plot(profile['TEMP2'], profile['PRES'],'g')
plt.gca().invert_yaxis()
plt.xlabel('Temperature [degC]')
plt.ylabel('Pressure [dbar]')
```

```python
# data
temp1 = S1['Temperature']
temp2 = S2['Temperature']
Sal1 = S1['Salinity']
Sal2 = S2['Salinity']
sp1 = S1['Sea pressure']
sp2 = S2['Sea pressure']
```

```python
#### Crie a figura e os eixos
fig, ax1 = plt.subplots(figsize=(8, 6))

# Plote a temperatura no eixo Y invertido
ax1.plot(temp1, sp1, 'r', label='Temperature 07-oct [degC]', linestyle='--')
ax1.plot(temp2, sp2, 'r', label='Temperature 11-oct [degC]')
ax1.set_xlabel('Temperature [degC]')
ax1.set_ylabel('Pressure [dbar]')
ax1.invert_yaxis()  # Inverta o eixo Y para que a pressão seja crescente para baixo

# Crie um segundo eixo compartilhando o mesmo eixo X
ax2 = ax1.twiny()

# Plote a salinidade no eixo X superior
ax2.plot(Sal1, sp1, 'g', label='Salinity 07-oct [PSU]', linestyle='--')
ax2.plot(Sal2, sp2, 'g', label='Salinity 11-oct [PSU]')
ax2.set_xlabel('Salinity [PSU]')

# Ajuste os limites dos eixos conforme necessário
ax1.set_xlim(14,17.5)  # Limite do eixo X inferior
ax2.set_xlim(32, 33)  # Limite do eixo X superior

# Adicione a legenda para a temperatura no canto inferior esquerdo
legend1 = ax1.legend(loc='lower left')

# Ajuste a posição da legenda para a salinidade
legend2 = ax2.legend(loc='upper left', bbox_to_anchor=(0.03, 0.23))  # Pode ajustar as coordenadas aqui

# Adicione um texto no gráfico
texto_x = 32.1 # Coordenada X do texto
texto_y = 4.5 # Coordenada Y do texto
texto = "Station 05"  # O texto que você deseja adicionar
plt.text(texto_x, texto_y, texto, fontsize=12, color='black')

plt.tight_layout()
plt.show()
```
